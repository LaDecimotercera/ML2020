```python
import numpy as np
import pandas as pd
from cvxopt  import solvers, matrix

# Training
X_train = pd.read_csv('./X_train.csv', header = None).values.tolist()
Y_train = pd.read_csv('./y_train.csv', header = None).values.tolist()

# Formatting
P = np.zeros([400,400]).tolist()
for i in range(400):
    for j in range(400):
        P[j][i] = Y_train[i][0]*Y_train[j][0]*np.dot(np.array(X_train[i]),np.array(X_train[j]))
P = matrix(P)
q = matrix(-np.ones((400,1)))
A = matrix(Y_train,(1,400),'d')
b = matrix(0.)
G = matrix(-np.identity(400))
h = matrix(np.zeros((400,1)))

sol = solvers.qp(P,q,G,h,A,b)
S_list = []
for idx in range(len(sol['x'])):
    if sol['x'][idx]>0: S_list.append(idx)

w = 0
for k in range(400):
    w += sol['x'][k]*Y_train[k][0]*np.array(X_train[k])

b = 0    
val = 0
for s in S_list:
    for i in S_list:
        val += sol['x'][i]*Y_train[i][0]*np.dot(np.array(X_train[i]),np.array(X_train[s])) 
    b += Y_train[s][0] - val
b = b/len(sol['x'])
```

```python
X_test = pd.read_csv('./X_test.csv', header = None).values
output = []
for i in range(len(X_test)):
    if np.dot(w,X_test[i])+b>=1: output.append(1)
    if np.dot(w,X_test[i])+b<=-1: output.append(-1)
#f(x) = w^Tx + b
print(output)
```
